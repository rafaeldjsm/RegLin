```
# -*- coding: utf-8 -*-
#Use to import pandas
import pandas as pd
import numpy as np
#Use to import the file into google Colab drive
from google.colab import files
df = pd.read_csv('/content/drive/My Drive/Dados.csv',encoding='latin1')
df=df.drop(columns=['Dado'])
dm=df.T
nl=dm.shape
nl = nl[1]

# Definir vetor/lista com:
vet = []; # define um vetor vazio
for i in range(0, nl, 1) : # ou mais simples neste caso "for i in range(n)"
  vet.append(int(1)); # anexe mais um elemento `a lista
# Imprime o vetor/lista de uma so' vez (na mesma linha)

vet=(pd.DataFrame(vet)).T

md=pd.concat([vet, dm])
mdt=md.T
mdt
```
